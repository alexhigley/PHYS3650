```python
#single gaussian

def line1(x,a1,mu,sig,c):
    y = 1 + a1*e**((-(x-mu)**2)/(2*sig**2))+c
    return y
```

```python
#five gaussians

def line2(x, a1,mu1,sig1, a2,mu2,sig2 ,a3,mu3,sig3, a4,mu4,sig4, a5,mu5,sig5,c):
    y1 = 1 + a1*e**((-(x-mu1)**2)/(2*sig1**2))
    y2 = 1 + a2*e**((-(x-mu2)**2)/(2*sig2**2))
    y3 = 1 + a3*e**((-(x-mu3)**2)/(2*sig3**2))
    y4 = 1 + a4*e**((-(x-mu4)**2)/(2*sig4**2))
    y5 = 1 + a5*e**((-(x-mu5)**2)/(2*sig5**2))
    
    y = y1 + y2 + y3 + y4 + y5 + c
    return y
```

```python
#four gaussians

def line3(x, a1,mu1,sig1, a2,mu2,sig2 ,a3,mu3,sig3, a4,mu4,sig4,c):
    y1 = 1 + a1*e**((-(x-mu1)**2)/(2*sig1**2))
    y2 = 1 + a2*e**((-(x-mu2)**2)/(2*sig2**2))
    y3 = 1 + a3*e**((-(x-mu3)**2)/(2*sig3**2))
    y4 = 1 + a4*e**((-(x-mu4)**2)/(2*sig4**2))
    
    y = y1 + y2 + y3 + y4 + c
    return y
```

```python
#three gaussians

def line4(x, a1,mu1,sig1, a2,mu2,sig2 ,a3,mu3,sig3,c):
    y1 = 1 + a1*e**((-(x-mu1)**2)/(2*sig1**2))
    y2 = 1 + a2*e**((-(x-mu2)**2)/(2*sig2**2))
    y3 = 1 + a3*e**((-(x-mu3)**2)/(2*sig3**2))
    
    y = y1 + y2 + y3 + c
    return y
```

```python
#single gaussian

def line5(x,a1,mu,sig):
    y = 1 + a1*e**((-(x-mu)**2)/(2*sig**2))
    return y
```

```python
# This function is specifically for experimenter 1 data. 
# Takes in current data, and a start and stop voltage to define a peak over
# Fits gaussian and finds center of defined peal

def getpeak(voltage,current,start,stop):
    
    loc = np.where((voltage >= start ) & (voltage <= stop))
    xdist = voltage[loc]
    ydist = current[loc]
    
    gg = np.mean(xdist)
    sd = np.std(xdist)

    p1= np.array([1,gg,sd,0])
    coeff1, covar_matrix1 = curve_fit(line1, xdist, ydist, p0=p1)
    xarr1=np.linspace(np.min(xdist),np.max(xdist),1000)
    yarr1=line1(xarr1,*coeff1)
    
    cent = coeff1[1]
    
    plt.plot(xdist,ydist)
    plt.plot(xarr1,yarr1)
    plt.xlabel('x');
    plt.ylabel('y');
    
    return cent
```
